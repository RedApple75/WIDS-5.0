```python
from PIL import Image
import numpy as np

def get_dims(file):
    im = Image.open(file)
    arr = np.array(im)
    h, w = arr.shape[:2]
    return h, w
```

```python
import os
import dask.bag as db
from dask.diagnostics import ProgressBar
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


for n, d in directories.items():
    filepath = d
    
    # FIX: use os.path.join instead of string concatenation
    filelist = [
        os.path.join(filepath, f)
        for f in os.listdir(filepath)
        if f.lower().endswith(('.jpg', '.jpeg', '.png'))
    ]
    
    dims = db.from_sequence(filelist).map(get_dims)
    
    with ProgressBar():
        dims = dims.compute()
    
    dim_df = pd.DataFrame(dims, columns=['height', 'width'])
    sizes = (
        dim_df
        .groupby(['height', 'width'])
        .size()
        .reset_index(name='count')
    )
    
    sizes.plot.scatter(x='width', y='height')
    plt.title(f'Image Sizes (pixels) | {n}')
    plt.xlabel('Width')
    plt.ylabel('Height')
    plt.show()
```

```python
BASE = r'F:\WIDS\data\plantvillage dataset'

categories = {
    'color': rf'{BASE}\color',
    'grayscale': rf'{BASE}\grayscale',
    'segmented': rf'{BASE}\segmented'
}
```

```python
from collections import defaultdict

dim_stats = {}

for cat, base_path in categories.items():
    filelist = []
    
    for cls in os.listdir(base_path):
        cls_path = os.path.join(base_path, cls)
        if os.path.isdir(cls_path):
            filelist += [os.path.join(cls_path, f) for f in os.listdir(cls_path)]
    
    dims = db.from_sequence(filelist).map(get_dims)

    with ProgressBar():
        dims = dims.compute()

    dim_df = pd.DataFrame(dims, columns=['height', 'width'])
    dim_stats[cat] = dim_df
```

```python
plt.figure(figsize=(15,5))

for i, (cat, df_) in enumerate(dim_stats.items()):
    plt.subplot(1,3,i+1)
    plt.scatter(df_['width'], df_['height'], s=1)
    plt.title(cat.capitalize())
    plt.xlabel('Width')
    plt.ylabel('Height')

plt.suptitle('Image Resolution Distribution Across Dataset Views')
plt.tight_layout()
plt.show()
```

```python
summary = []

for cat, df_ in dim_stats.items():
    summary.append({
        'category': cat,
        'unique_resolutions': df_.drop_duplicates().shape[0],
        'mean_height': df_['height'].mean(),
        'mean_width': df_['width'].mean(),
        'std_height': df_['height'].std(),
        'std_width': df_['width'].std()
    })

pd.DataFrame(summary)
```

### Resolution Consistency Across Views
Color and grayscale images are perfectly resolution-aligned (256Ã—256). Segmented images preserve nominal size but show minor geometric variability due to cropping and padding, indicating segmentation alters information density rather than spatial scale.


```python
from PIL import Image
import os

bad_images = []

for cls, path in directories.items():
    for f in os.listdir(path)[::]:  # sample
        try:
            Image.open(os.path.join(path, f)).verify()
        except:
            bad_images.append((cls, f))

print("Corrupt images found:", len(bad_images))
```

```python
import cv2
import numpy as np

def blur_score(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    return cv2.Laplacian(img, cv2.CV_64F).var()

blur_scores = []

for cls, path in list(directories.items())[::]:  # sample classes
    for f in os.listdir(path)[::]:
        blur_scores.append(blur_score(os.path.join(path, f)))

np.mean(blur_scores), np.std(blur_scores)
```

```python
import matplotlib.pyplot as plt
plt.hist(blur_scores, bins=50)
plt.title("Image Sharpness Distribution")
plt.xlabel("Laplacian Variance")
plt.ylabel("Frequency")
plt.show()
```

Blur analysis shows a right-skewed distribution: most images are sharp, with a small number extremely sharp samples. This reflects controlled acquisition conditions and suggests that models may overfit to high-frequency texture unless augmented.

```python
def non_black_ratio(img_path):
    img = np.array(Image.open(img_path))
    return np.count_nonzero(img) / img.size

ratios = []

seg_base = r'F:\WIDS\data\plantvillage dataset\segmented'

for cls in os.listdir(seg_base):
    cls_path = os.path.join(seg_base, cls)
    for f in os.listdir(cls_path)[::]:
        ratios.append(non_black_ratio(os.path.join(cls_path, f)))

np.mean(ratios), np.std(ratios)
```

```python
def rgb_variance(img_path):
    img = np.array(Image.open(img_path))
    return img[:,:,0].var(), img[:,:,1].var(), img[:,:,2].var()

variances = []

for cls in ['Apple___healthy', 'Apple___Black_rot']:
    path = directories[cls]
    for f in os.listdir(path)[::]:
        variances.append(rgb_variance(os.path.join(path, f)))

np.mean(variances, axis=0)
```

- Dataset contains zero corrupted images.
- Blur scores are right-skewed, indicating predominantly sharp images captured under controlled conditions.
- Segmentation removes ~47% of pixels on average, changing information density rather than resolution.
- RGB channels exhibit unequal variance (Blue > Red > Green), indicating color-dependent disease cues.
